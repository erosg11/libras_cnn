```python
with open('full/curve_ufpr v2.json') as fp:
    data_curves = load(fp)
```

```python
from IPython.display import JSON
JSON(data_curves)
```

```python
# Essa parte ficou confusa pra caranba, pode me xingar
# Como pode ver acima, ele está no formato {Algoritmo: {K: {métricas: [valores]}}}
dfs = {}
for algoritmo, k_fold_dado in data_curves.items():
    data = {}
    for indice_k, dado_metrica in k_fold_dado.items():
        for metrica, valores in dado_metrica.items():
            data[(indice_k, metrica)] = valores
    dfs[algoritmo] = pd.DataFrame(data)
    dfs[algoritmo].loc[:, (dfs[algoritmo].columns.get_level_values(1).isin(['accuracy', 'val_accuracy']))] *= 100
    display(dfs[algoritmo].head())
```

```python
# Aplicando métricas de agrupamento dos dados
for k, df in dfs.items():
    for name, metric in [('mean', 'mean')]:
        for item in df.columns.get_level_values(1).unique():
            dfs[k][(name, item)] = df.loc[:, df.columns.get_level_values(1)==item].apply(metric, axis=1)
    print(k)
    display(df.head())
```

```python
def create_plot(dfs, item, mean_metric='median', paletas=None):
    """
    Função para gerar a vizualização dos dados
    param dfs: Dicionário de dataframes, ele que contem os dados a serem percorridos
    param item: Métrica a ser plotada
    param mean_metric: Qual métrica de agrupamento será usada para mensurar o valor médio dos dados
    param paletas: Dicionário com as chaves sendo iguais ao do dfs e valor sendo uma paleta com no mínimo 3 cores
    returns: Gráfico formado pelos parâmetros fornecidos
    """
    dfs = dfs.copy()
    paletas = paletas or {'LeNet': palettes.Blues3, 'VGG19': palettes.Greens3, 'InceptionV3': palettes.Oranges3, 'ResNet50': palettes.Greys3, 'VGG16': palettes.BuPu3}
    lines = {}
    for k, df in dfs.items():
        dfs[k] = dfs[k].copy()
        dfs[k] = df.loc[:, (df.columns.get_level_values(1)==item) & (df.columns.get_level_values(0) == mean_metric)]
        lines[k] = hv.Curve(dfs[k].loc[:, dfs[k].columns.get_level_values(0)==mean_metric].values[:,0], ('x', 'Época'),
                                                        ('y', 'Custo' if item.endswith('loss') else 'Acurácia')).opts(tools=['hover', 'ywheel_zoom', 'xwheel_pan'], 
                                                                                                             default_tools=['save', 'pan', 'box_zoom', 'reset'], color=paletas[k][0], muted_alpha=0.2)
        
    return hv.NdOverlay(lines).opts(width=int(800//2), height=int(500//2), legend_position='top_right' if item.endswith('loss') else 'bottom_right')
```

```python
plot = reduce(add, [create_plot(dfs, metric, mean_metric='mean').relabel(metric.capitalize()) for metric in df.columns.get_level_values(1).unique()]).cols(1).opts(shared_axes=False)
plot
```

```python
hv.save(plot, 'dashboard UFPR full.html')
```
